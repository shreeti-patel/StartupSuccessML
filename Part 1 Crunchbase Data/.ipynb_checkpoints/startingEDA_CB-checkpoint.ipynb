```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data from CSV files ##

```python

# import pandas as pd

# Load datasets for 2014
companies_2014 = pd.read_csv('2014-crunchbase-data/companies_2014.csv')
rounds_2014 = pd.read_csv('2014-crunchbase-data/rounds_2014.csv')
investments_2014 = pd.read_csv('2014-crunchbase-data/investments_2014.csv')
acquisitions_2014 = pd.read_csv('2014-crunchbase-data/aquisitions_2014.csv')

# Load datasets for 2015
companies_2015 = pd.read_csv('2015-crunchbase-data/companies_2015.csv')
rounds_2015 = pd.read_csv('2015-crunchbase-data/rounds_2015.csv')
investments_2015 = pd.read_csv('2015-crunchbase-data/investments_2015.csv')
acquisitions_2015 = pd.read_csv('2015-crunchbase-data/acquisitions_2015.csv')

print(companies_2014.columns)
print(companies_2015.columns)
```

## Rename Columns in 2014 Data ##

```python
# Fix whitespaces in column names
print(companies_2014.columns)
column_mapping_2014 = {
    ' market ': 'market',
    ' funding_total_usd ' : '2014_funding_total_usd',
    'status' : '2014_status',
    'funding_rounds': '2014_funding_rounds',
    'last_funding_at': '2014_last_funding_at'
}


# Use the rename method to apply the renaming
companies_2014 = companies_2014.rename(columns=column_mapping_2014)
companies_2014.columns
```

## Merge relavant 2014 with 2015 data ##

```python
import pandas as pd

# Merge df_2015 and df_2014 based on 'permalink'
merged_df = pd.merge(companies_2015, companies_2014[['permalink', '2014_funding_total_usd', '2014_funding_rounds', '2014_status', '2014_last_funding_at']], on='permalink', how='left')
merged_df

# To update df_2015 in place:
# companies_2015[['2014_funding_total_usd', '2014_funding_rounds', '2014_last_funding_at']] = merged_df[['2014_funding_total_usd', '2014_funding_rounds', '2014_last_funding_at']]

```

```python
merged_df.columns
```

## Handle missing values and clean data ##

```python
#Clean Data

# Drop companies outside USA
df = merged_df[merged_df['country_code'] == 'USA']

# # Drop companies that do not have both 2014 and 2015 data
# df = df.dropna(subset=['2014_funding_total_usd','2014_funding_rounds', '2014_status', '2014_last_funding_at'])
# df

# Removing duplicate rows based on all columns
df = df.drop_duplicates()
  
df
```

## Convert Data Types ##

```python

# Convert date columns to datetime data type with custom date range
df['founded_at'] = pd.to_datetime(df['founded_at'], errors='coerce', infer_datetime_format=True, utc=True)
df['first_funding_at'] = pd.to_datetime(df['first_funding_at'], errors='coerce', infer_datetime_format=True, utc=True)
df['last_funding_at'] = pd.to_datetime(df['last_funding_at'], errors='coerce', infer_datetime_format=True, utc=True)
df['2014_last_funding_at'] = pd.to_datetime(df['2014_last_funding_at'], errors='coerce', infer_datetime_format=True, utc=True)

#df.to_csv('out.csv')
df.info()
df
```

### 2014_funding_rounds - Drop Missing Values and Convert Data Type

```python
# # Example for converting funding-related columns to numeric data type
#df[['funding_total_usd', '2014_funding_total_usd']] = df[['funding_total_usd', '2014_funding_total_usd']].apply(pd.to_numeric, errors='coerce')

# # Convert '2014_funding_rounds' column to float with errors='coerce'
# df['2014_funding_rounds'] = df['2014_funding_rounds'].astype(int, errors='coerce')
# Convert '2014_funding_rounds' column to float with errors='coerce'
#df['2014_funding_rounds'] = pd.to_numeric(df['2014_funding_rounds'], errors='coerce')

# # Fill NaN values with 0 in '2014_funding_rounds' column
df['2014_funding_rounds'] = df['2014_funding_rounds'].fillna(0)

# # Convert '2014_funding_rounds' column to integer
df['2014_funding_rounds'] = df['2014_funding_rounds'].astype(int)

df.info()
```

### 2014_funding_total_usd - Convert to numeric

```python

df['2014_funding_total_usd'] = df['2014_funding_total_usd'].str.replace(' -   ', '')

df['2014_funding_total_usd'] = df['2014_funding_total_usd'].str.replace(',', '')
df['2014_funding_total_usd'] = pd.to_numeric(df['2014_funding_total_usd'], errors='coerce')


df['funding_total_usd'] = df['funding_total_usd'].str.replace(' -   ', '')

df['funding_total_usd'] = df['funding_total_usd'].str.replace(',', '')
df['funding_total_usd'] = pd.to_numeric(df['funding_total_usd'], errors='coerce')

df.info()
```

## Drop rows not in chronological order 
#### Some values seem to have the first funding date befoer hte founded at date. Lets drop any columns where founded_at , first_funding and last_funding are not in chronological order

```python

# Identify rows where the chronological order is not met
rows_not_in_order = df[~(df['founded_at'] <= df['first_funding_at']) | 
                       ~(df['first_funding_at'] <= df['last_funding_at'])]

# Drop the rows not in chronological order
df = df.drop(rows_not_in_order.index)

# Optionally, reset the index after dropping rows
df.reset_index(drop=True, inplace=True)
df
```

### All companies where 'acquirer_name' is not null and 'status' is not 'acquired' have status as closed. This means they were likely aquired and then closed. Should these companies be marked as sucessful startups? Yes, if they were aquired, they are sucessful.

## Defining Success
#### Let's define if a start up is sucessful or not by creating the is_success column. If the company's status is ipo or aquired or if acquirer_name is not NaN ()meaning it has been aquired) , it is considered a succesful start up

```python
acq_df['is_success'] = acq_df.apply(lambda row: 1 if (row['status'] == 'ipo') or (row['status'] == 'acquired') or pd.notna(row['acquirer_name']) else 0, axis=1)
acq_df.info()
```

### Dataframe of the successful startups

```python
succesful_startups = acq_df[acq_df['is_success'] == 1]
succesful_startups
```

```python
sns.countplot(x='is_success', data=acq_df)
plt.show()
```
