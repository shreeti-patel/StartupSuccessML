## Continued Category List Feature Engineering

### Dataframe of just the different categories present within all companies

```python
 
new_df = df_exploded['category_list'].copy()
new_df = pd.DataFrame(df_exploded['category_list'].values, columns=['category'])

new_df
```

### Use TF-IDF (Term Frequency-Inverse Document Frequency) vectorization and K-Means clustering algorithms from the scikit-learn library 

TF-IDF is a numerical stat that reflects how important a word is to a document in a collection



```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Vectorize category names
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(new_df['category'])

# Cluster categories (assuming you want to reduce to 6 categories)
kmeans = KMeans(n_clusters=60, random_state=0).fit(X)

# Assign the cluster labels as the new categories
new_df['new_category'] = kmeans.labels_
```

```python
new_df
```

### Visualize the value count of the new category groups

```python
# Ensure you're using the exploded DataFrame where each category is in its own row
plt.figure(figsize=[8, 10])
sns.countplot(data=new_df, y='new_category', order=new_df['new_category'].value_counts().index)
plt.title('Count Plot of Categories')
plt.xlabel('Counts')
plt.ylabel('Categories')
plt.show()
```
